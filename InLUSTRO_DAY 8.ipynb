## NAIVE BAYES

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = pd.read_csv('dataset/app_review.csv')
data.head()
```

```python
def preprocess_data(data):
    # Remove package name as it's not relevant
    data = data.drop('package_name', axis=1)
    # convrt txt to lowercase
    # strip - removes whitesapces in first or last spaces in data
    data['review']=data['review'].str.strip().str.lower()
    return data
data = preprocess_data(data)
data.head()
```

```python
#split into training and testing
x = data['review'] 
y = data['polarity']
x,x_test,y,y_test = train_test_split(x,y, stratify=y, test_size=0.25,random_state=42)

# stratify - for balancing the training and testing
```

```python
# Vectorize text reviews to numbersa
vec = CountVectorizer(stop_words='english')
x = vec.fit_transform(x).toarray()
x_test = vec.transform(x_test).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x,y)
```

```python
model.score(x_test,y_test)
```

### SENTIMENT ANALYSIS classifiaction

```python
model.predict(vec.transform(['Love this app simply awesome!']))
```

```python
model.predict(vec.transform(["It's getting stuck a lot while using this app"]))
```

```python
print(model.score(x_test,y_test))
print(model.predict(vec.transform(['Love this app simply awesome!'])))
print(model.predict(vec.transform(["It's getting stuck a lot while using this app"])))
```

## DECISION TREES

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier# import Decision TreeClassier
from sklearn.model_selection import train_test_split #import train_test_split
from sklearn import metrics # accuracy calculator
```

```python
col_names = ['pregnant','glucose','bp','skin','insulin','bmi','pedigree','age','label']
#load dataset
pima = pd.read_csv("dataset/pima-indians-diabetes.csv",header=None,names=col_names)
pima.head()
```

```python
feature_cols = ['pregnant','insulin','bmi','age','glucose','bp','pedigree']
x=pima[feature_cols ] # features
y=pima.label # target variable
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)  # 70% training and 30% testing
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)
clf = clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('dataset/User_Data.csv')
x=data.iloc[:,[2,3]].values
y=data.iloc[:,4].values
print(x[:10])
```

```python
#extracting independent and dependent variable
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc_x= StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
print(x_train[:10])
x_test=sc_x.transform(x_test)
```

```python
# Fitting K-NN classifier to the training set
# metrics - minkowski,euclidean distance,manhattan

from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5, metric='minkowski',p=2)
classifier.fit(x_train, y_train)
```

```python
# predicting the test set result
y_pred=classifier.predict(x_test)
```

```python
#Creating the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y_test,y_pred))
      
```

```python
from sklearn.metrics import precision_score
print("Precision:",precision_score(y_test,y_pred))
```

```python
# MANHATTAN
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5, metric='manhattan',p=2)
classifier.fit(x_train, y_train)
```

```python
# EUCLIDEAN
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5, metric='euclidean',p=2)
classifier.fit(x_train, y_train)
```
